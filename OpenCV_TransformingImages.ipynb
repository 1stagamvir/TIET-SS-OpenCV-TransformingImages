```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Import cv2_imshow
# Load our image
image = cv2.imread('/content/logo.jpg')

# Add a check to see if the image was loaded
if image is not None:
    cv2_imshow(image)

    # Store height and width of the image
    height, width = image.shape[:2]

    # We shift it by quarter of the height and width
    quarter_height, quarter_width = height/4, width/4

    # Our Translation
    #       | 1 0 Tx |
    #  T  = | 0 1 Ty |

    # T is our translation matrix
    T = np.float32([[1, 0, quarter_width], [0, 1,quarter_height]])

    # We use warpAffine to transform the image using the matrix, T
    img_translation = cv2.warpAffine(image, T, (width, height))
    cv2_imshow(img_translation)
else:
    print("Could not load the image 'images/flowers.jpg'. Please check the path and file.")
```

```python
# Load our image
image = cv2.imread('/content/logo.jpg')
height, width = image.shape[:2]

# Divide by two to rototate the image around its centre
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)

# Input our image, the rotation matrix and our desired final width and height
rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
cv2_imshow(rotated_image)
```
